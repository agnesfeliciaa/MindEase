**Membuat table missing values**

```python
#create dataframe to summarize missing values
def missing_value_pct_df(data):
  percent_missing =data.isnull().mean()
  missing_value_df=pd.DataFrame (percent_missing).reset_index()
  missing_value_df=missing_value_df.rename(columns={"index":"column_name", 0: "percent_missing"})
  #multiply by 100 and round to 4 decimal places
  missing_value_df["percent_missing"]=missing_value_df["percent_missing"].apply(lambda x: round(x*100, 2))
  missing_value_df=missing_value_df.sort_values(by=["percent_missing"], ascending=False)
  return missing_value_df
```

### Remove Outliner

```python
outliers_index = []
for column in df:
    # Calculate IQR for price
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Identify outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    for idx in outliers.index:
        if idx not in outliers_index:
            outliers_index.append(idx)

    print(f"Number of outliers in {column}: {len(outliers)}")
```

**keterangan:** Kode diatas digunakan untuk melakukan remove outlier untuk data yang memiliki kecenderungan. Karena *mengidentifikasi* dan menangani outlier adalah bagian penting dari proses ini.

```python
# import matplotlib.pyplot as plt
# import seaborn as sns

# # Visualize outliers for 'noise_level' column
# plt.figure(figsize=(10, 6))
# sns.boxplot(x=df['noise_level'])
# plt.title('Noise Level Boxplot')
# plt.show()
```

```python
lower_bound = Q1 - 2.0 * IQR  # Adjust multiplier
upper_bound = Q3 + 2.0 * IQR  # Adjust multiplier
for column in df.columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Replace outliers with the median
    df[column] = df[column].apply(lambda x: df[column].median() if x < lower_bound or x > upper_bound else x)
```

```python
import pandas as pd

# Daftar untuk menyimpan indeks outlier
outliers_index = []

for column in df:
    # Pastikan kolom tersebut bertipe numerik
    if df[column].dtype in ['int64', 'float64']:
        # Menghitung IQR (Interquartile Range)
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        # Menentukan batas bawah dan atas
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Identifikasi outlier
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

        # Menyimpan indeks outlier
        for idx in outliers.index:
            if idx not in outliers_index:
                outliers_index.append(idx)

        print(f"Number of outliers in {column}: {len(outliers)}")

        # Menangani outlier (mengganti dengan nilai batas bawah atau atas)
        df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)

        # Mengonversi kolom kembali ke tipe integer
        if df[column].dtype == 'float64':
            df[column] = df[column].astype('int')

# Pastikan tipe data tetap integer setelah perbaikan
print(df.dtypes)
```

# **Modelling/Model Training**

```python
# Pisahkan fitur dan target dari dataset
X = df.drop(columns=['stress_level'])  # Fitur-fitur
y = df['stress_level']  # Target
```

**Keterangan:** Dalam langkah ini, membagi dataset menjadi dua bagian: satu untuk pelatihan (training) dan satu lagi untuk pengujian (testing). Pembagian ini penting untuk memastikan bahwa model dapat dilatih pada satu subset data dan dievaluasi pada subset yang berbeda, sehingga mencegah overfitting dan memberikan gambaran akurat tentang performa model di data yang belum pernah dilihat sebelumnya.
Digunakan untuk membagi dataset menjadi dua subset: data latih dan data uji, menggunakan fungsi train_test_split dari pustaka scikit-learn. Dengan test_size=0.2, 20% dari data akan dialokasikan untuk pengujian, sementara 80% sisanya akan digunakan untuk melatih model. Parameter random_state=0 memastikan bahwa pembagian ini dapat direproduksi, sehingga hasilnya konsisten setiap kali dijalankan.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Hitung Z-Score untuk y_test dan y_pred
z_scores = np.abs(zscore(y_test - y_pred))
outlier_threshold = 3  # Threshold untuk mendeteksi outlier (standar: 3)

# Identifikasi Outlier
outlier_indices = np.where(z_scores > outlier_threshold)[0]

# Plot Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, label='Data Points', alpha=0.7, color='blue')
plt.scatter(
    y_test.iloc[outlier_indices],
    y_pred[outlier_indices],
    label='Outliers',
    color='red',
    edgecolor='k',
    s=100
)

plt.xlabel("Y Test")
plt.ylabel("Y Predicted")
plt.title("Scatter Plot of Predictions with Outliers Highlighted")
plt.plot(y_test, y_test, c='r', label='Ideal Fit Line')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Hapus outlier
X_test_clean = X_test.iloc[~y_test.index.isin(outlier_indices)]
y_test_clean = y_test.iloc[~y_test.index.isin(outlier_indices)]

# Prediksi ulang tanpa outlier
y_pred_clean = lr.predict(X_test_clean)

# Evaluasi ulang
plt.figure(figsize=(10, 6))
plt.scatter(y_test_clean, y_pred_clean, label='Cleaned Data', color='blue')
plt.plot(y_test_clean, y_test_clean, c='r', label='Ideal Fit Line')
plt.xlabel("Y Test (Cleaned)")
plt.ylabel("Y Predicted (Cleaned)")
plt.title("Scatter Plot Without Outliers")
plt.legend()
plt.grid(True)
plt.show()
```

# **Evaluation**

**Evaluasi kualitas model dengan menggunakan berbagai evaluastion metrics**

```python
#Evaluasi
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
import numpy as np
mae=mean_absolute_error(y_test, y_pred)
mse=mean_squared_error(y_test, y_pred)
rmse=np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
print(f"Mean Absolute Error: {mae:.2f}")
print(f"Mean Squared Error: {mse:.2f}")
print(f"Root Mean Squared Error: {rmse:.2f}")
print(f"R2 score: {r2:.2f}")
```

**Keterangan :**
* mae = mean_absolute_error(y_test, y_pred) →MAE mengukur rata-rata kesalahan absolut antara nilai sebenarnya dan nilai prediksi untuk memberikan gambaran yang jelas tentang seberapa besar kesalahan prediksi model dalam unit yang sama dengan variabel dependen (stress_level).  
* mse = mean_squared_error(y_test, y_pred)--> MSE mengukur rata-rata kuadrat dari kesalahan prediksi. Semakin kecil nilai MSE, semakin baik kualitas model tersebut. Metrik ini lebih sensitif terhadap outlier karena kesalahan kuadratnya lebih besar.  
* rmse = np.sqrt(mse) adalah akar kuadrat dari MSE dan memberikan kesalahan dalam unit yang sama dengan variabel dependen. Semakin kecil nilai RMSE, maka semakin baik kualitas model. RMSE sering dianggap sebagai metrik yang lebih intuitif karena lebih mudah diinterpretasikan. RMSE rendah menunjukkan kinerja model yang lebih baik.
* r2 = r2_score(y_test, y_pred): R² score mengukur seberapa baik model menjelaskan variasi dalam data. Nilai R² berkisar antara 0 hingga 1, dengan nilai lebih tinggi menunjukkan kinerja model yang lebih baik. Semakin tinggi nilai R2, maka semakin baik kinerja model dalam menjelaskan hubungan antara variabel. Nilai mendekati 1 menunjukkan model yang baik.





**Melihat error yang ada pada MEAN**

```python
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

mape_train = mean_absolute_percentage_error(y_train, y_predict_train)
mape_train
```
